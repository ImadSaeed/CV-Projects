```python
import os
import sys
from pathlib import Path
from google.colab import drive
from IPython.display import Image, display

# 1. Mount Drive (Stable Connection)
drive.mount('/content/drive', force_remount=True)

# 2. Paths Configuration
# Source: Google Drive
DRIVE_ZIP = "/content/drive/MyDrive/PlantVillage_Project/data/raw/color.zip"
# Destination: The high-speed local SSD
LOCAL_DATA_ROOT = "/content/data/raw"
LOCAL_EXTRACT_PATH = "/content/data/raw/color"

# 3. Clean and Unzip from Drive to SSD
if os.path.exists(DRIVE_ZIP):
    print(" Cleaning local SSD for fresh start...")
    !rm -rf {LOCAL_DATA_ROOT}
    !mkdir -p {LOCAL_DATA_ROOT}

    print("Unzipping from Drive to Local SSD (Fast & Stable)...")
    # Extracting directly to /content/data/raw
    !unzip -q "{DRIVE_ZIP}" -d {LOCAL_DATA_ROOT}

    # 4. Verification of the 38 Folders
    if os.path.exists(LOCAL_EXTRACT_PATH):
        folders = [f for f in os.listdir(LOCAL_EXTRACT_PATH) if os.path.isdir(os.path.join(LOCAL_EXTRACT_PATH, f))]
        print(f"SUCCESS! Local SSD count: {len(folders)}")

        if len(folders) < 38:
            print(f"Warning: Still only found {len(folders)}. This confirms the ZIP on Drive is incomplete.")
        else:
            # 5. Point Training to the Local SSD
            PROJECT_ROOT = "/content/drive/MyDrive/PlantVillage_Project"
            os.environ["RAW_DATA_PATH"] = LOCAL_EXTRACT_PATH
            sys.path.append(os.path.join(PROJECT_ROOT, "code"))
            sys.path.append(os.path.join(PROJECT_ROOT, "code/modules"))

            # 6. Start Phase 1 Training on T4 GPU
            print("\nLaunching train_head.py using Local SSD...")
            try:
                os.chdir(PROJECT_ROOT)
                %run -i "code/scripts/train_head.py"

                # Show results if completed
                curve_path = os.path.join(PROJECT_ROOT, "evaluation/tiny_nn/phase1_training_curves.png")
                if os.path.exists(curve_path):
                    display(Image(filename=curve_path))
            except Exception as e:
                print(f"Training failed: {e}")
    else:
        print(f"ERROR: Extraction failed. {LOCAL_EXTRACT_PATH} not created.")
else:
    print(f"ERROR: Zip not found on Drive at {DRIVE_ZIP}")
```
